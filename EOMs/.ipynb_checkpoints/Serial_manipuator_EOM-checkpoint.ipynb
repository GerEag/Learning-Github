# 2-link serial manipulator

Using kane's method to derive equations of motion for a 2-link serial manipulator

NOTE: Potential problem with joint torques

```python
from sympy import *
from sympy.physics.mechanics import *
from pydy.system import System
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
%matplotlib inline

# define generalized coordinates
th1, th2 = dynamicsymbols('th1 th2')
th1d, th2d = dynamicsymbols('th1 th2', 1)

# define generalized velocities
u1, u2 = dynamicsymbols('u1 u2')
u1d, u2d = dynamicsymbols('u1 u2', 1)

# define control torques
tau1, tau2 = dynamicsymbols('tau1 tau2')

# define costant parameters
m1, m2, L1, L2, g = symbols('m1 m2 L1 L2 g')

# define reference frames
N = ReferenceFrame('N') # Fixed frame
C1 = N.orientnew('C1', 'Axis', [th1, N.z]) # frame along cable 1
C2 = N.orientnew('C2', 'Axis', [th2, N.z]) # frame along cable 2

# set angular velocity of frames
C1.set_ang_vel(N, u1 * N.z) # cable 1 rotates at ang vel. of u1 about N.z
C2.set_ang_vel(N, u2 * N.z) # cable 2 rotates at ang. vel. of u2 about N.z

# define points
T = Point('T') # fixed point
P1 = T.locatenew('P1', L1 * C1.x) # location of link 1 mass (also defines axis from which th1 is measured)
P2 = P1.locatenew('P2', L2 * C2.x) # location of link 2 mass (also defines axis from which th2 is measured)

# define velocities
T.set_vel(N, 0 * N.x)
P1.v2pt_theory(T, N, C1)
P2.v2pt_theory(P1, N, C2)

# define bodies and particles
ParP1 = Particle('ParP1', P1, m1)
ParP2 = Particle('ParP2', P2, m2)

# kinodynamic constraints
kd = [th1d - u1, th2d - u2]

# define forces and moments
FL = [(P1, - m1 * g * N.y),
      (P2, - m2 * g * N.y),
      (C1, tau1 * C1.z),
      (C2, tau2 * C2.z)]

# define bodies
BL = [ParP1, ParP2]

KM = KanesMethod(N, q_ind=[th1, th2], u_ind=[u1, u2], kd_eqs=kd)
(fr, frstar) = KM.kanes_equations(BL,FL)

# mechanics_printing(pretty_print=True)

# KM.rhs()
```

```python
sys = System(KM)
sys.constants = {
                 m1:1,
                 m2:100,
                 L1:2,
                 L2:1,
                 g:9.81,
                 }
sys.initial_conditions = {th1:-np.pi/2, th2:-np.pi/2,u1:0, u2:0}
sys.specifieds = {tau1: control_tau1, tau2:control_tau2}
sim_time = np.linspace(0.0, 5.0, 5000)
sys.times = sim_time
sys.generate_ode_function(generator='lambdify')

y = sys.integrate()
```

```python
# fig = plt.figure(figsize=(18, 4))
fig = plt.figure(figsize=(12,8))

# fig = plt.figure(0)
fig.add_subplot(221)
plt.plot(sim_time, y[:,0], label='Unshaped')
# plt.ylim(25,0)
plt.title(r'Link 1 Displacement')

fig.add_subplot(222)
plt.plot(sim_time, y[:,1], label='Unshaped')
# plt.gca().invert_yaxis()
# plt.ylim(20,0)
plt.title(r'Link 2 Displacement')

fig.add_subplot(223)
plt.plot(sim_time, y[:,2], label='Unshaped')
plt.title(r'Link 1 Velocity')

fig.add_subplot(224)
plt.plot(sim_time, y[:,3], label='Unshaped')
plt.title(r'Link 2 Velocity')

# plt.tight_layout()
plt.show()
```
